### **PASO 1**

- Creamos un modelo determinístico, es decir, sin incertidumbre.

```python
item1 = 100
item2 = 200
item3 = 150

try:
    presupuesto_total = item1 + item2 + item3

    print(f"El presupuesto total es de {presupuesto_total} usd")

except:
    print('ERROR')

print('\nFin del programa')
```

### **PASO 2**

- Ahora vamos a introducir incertidumbre

```python
import numpy as np

try:
    item1 = np.random.normal(100, 10)  # media 100, desviación estándar 10
    item2 = np.random.uniform(180, 220)  # uniforme entre 180 y 220
    item3 = np.random.triangular(100, 150, 200)  # triangular

    print('Programa exitoso')

except:
    print('ERROR')
```

### **PASO 3**

- Simulación con 5000 iteraciones

```python
presupuestos = []

try:
    for x in range(5000):
        item1 = np.random.normal(100, 10)
        item2 = np.random.uniform(180, 220)
        item3 = np.random.triangular(100, 150, 200)

        total = item1 + item2 + item3
        presupuestos.append(total)
    
    print('Simulación exitosa')
except:
    print('ERROR')
```

### **PASO 4**

- Interpretación de resultados

```python
import pandas as pd

try:
    resultados = pd.Series(presupuestos)
    print("Media:", resultados.mean())
    print("P95:", resultados.quantile(0.95))
    print("P5:", resultados.quantile(0.05))

    print('\nPrograma exitoso')
except:
    print('ERROR')
```

### **PASO 5**

- Visualización básica

```python
import matplotlib.pyplot as plt

plt.hist(presupuestos, bins=50, color='skyblue')
plt.title("Distribución del presupuesto total")
plt.xlabel("Costo total")
plt.ylabel("Frecuencia")
plt.show()
```

### **PASO 6**

- Análisis de sensiblidad, vamos a cambiar una de las distribuciones y vemos como afecta el resultado.

```python
presupuestos2= []

try:
    for x in range(5000):
        #item1 = np.random.normal(100, 10) # está fue la distribución anterior
        item1 = np.random.uniform(80, 120) 
        item2 = np.random.uniform(180, 220)
        item3 = np.random.triangular(100, 150, 200)

        total = item1 + item2 + item3
        presupuestos2.append(total)

    plt.hist(presupuestos2, bins=50, color='skyblue')
    plt.title("Distribución del presupuesto total")
    plt.xlabel("Costo total")
    plt.ylabel("Frecuencia")
    plt.show()
    
    print('Simulación exitosa')
except:
    print('ERROR')
```

```python
try:
    resultados = pd.Series(presupuestos)
    print("Resultados de la primera simulación:")
    print("Media:", resultados.mean())
    print("P95:", resultados.quantile(0.95))
    print("P5:", resultados.quantile(0.05))
    print("\nResultados de la segunda simulación:")
    resultados2 = pd.Series(presupuestos2)
    print("Media:", resultados2.mean())
    print("P95:", resultados2.quantile(0.95))
    print("P5:", resultados2.quantile(0.05))



    print('\nPrograma exitoso')
except:
    print('ERROR')

print('\nFin del programa')
```
